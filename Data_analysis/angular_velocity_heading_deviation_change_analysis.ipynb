# Angular Velocity-Based Heading Deviation CHANGE Analysis

This notebook analyzes the **change (rate) in heading deviation** as a function of **continuous angular velocity** (turning rate). This examines whether turning relates to error accumulation/correction rates rather than absolute error.

## Key Differences from Original Analysis:
- **Original**: Correlates angular velocity with absolute heading deviation
- **This notebook**: Correlates angular velocity with CHANGE in heading deviation
- **Biological meaning**: Does turning predict the rate of error correction/accumulation?

## Analysis Pipeline:
1. Calculate instantaneous heading from position data
2. Compute angular velocity (dθ/dt) at each timestep
3. Compute change in heading deviation (d(error)/dt) at each timestep
4. Bin data by angular velocity magnitude and direction
5. Analyze change in heading deviation as a function of turning rate
6. Compare left vs right turning and light vs dark conditions

## Data Source:
- Uses reconstruction data from `reconstuctionDFAutoPI.csv`
- Analyzes `mvtDirError` (heading deviation) and position (x, y)

## Calculation Functions

```python
def calculate_heading_from_position(x, y, smooth_window=0.1):
    """Calculate instantaneous heading from position data."""
    if smooth_window > 1:
        x_smooth = scipy.ndimage.gaussian_filter1d(x, sigma=smooth_window/3)
        y_smooth = scipy.ndimage.gaussian_filter1d(y, sigma=smooth_window/3)
    else:
        x_smooth, y_smooth = x, y
    
    dx = np.diff(x_smooth)
    dy = np.diff(y_smooth)
    heading = np.full(len(x), np.nan)
    heading[1:] = np.arctan2(dy, dx)
    return heading

def calculate_angular_velocity(heading, time, smooth_window=1.):
    """Calculate angular velocity from heading and time."""
    dheading = np.diff(heading)
    dheading = np.arctan2(np.sin(dheading), np.cos(dheading))
    dt = np.diff(time)
    dt[dt == 0] = np.nan
    angular_vel = dheading / dt
    angular_velocity = np.full(len(heading), np.nan)
    angular_velocity[1:] = angular_vel
    
    if smooth_window > 0:
        valid_mask = ~np.isnan(angular_velocity)
        if np.sum(valid_mask) > 0:
            angular_velocity[valid_mask] = scipy.ndimage.gaussian_filter1d(
                angular_velocity[valid_mask], sigma=smooth_window)
    return angular_velocity

def calculate_speed(x, y, time):
    """Calculate instantaneous speed from position and time data."""
    dx = np.diff(x)
    dy = np.diff(y)
    dt = np.diff(time)
    dt[dt == 0] = np.nan
    distance = np.sqrt(dx**2 + dy**2)
    speed_vals = distance / dt
    speed = np.full(len(x), np.nan)
    speed[1:] = speed_vals
    return speed

def calculate_heading_deviation_change(heading_deviation, time, smooth_window=0.5):
    """
    Calculate the change (derivative) in heading deviation over time.
    
    This represents the rate at which navigation error is accumulating or
    being corrected. Like angular velocity, this is a linear variable representing
    a rate of change in an angular quantity.
    
    Parameters
    ----------
    heading_deviation : array-like
        Heading deviation in radians (circular variable)
    time : array-like
        Time values
    smooth_window : float
        Smoothing window for Gaussian filter (0 for no smoothing)
    
    Returns
    -------
    heading_dev_change : ndarray
        Rate of change in heading deviation (rad/s)
    """
    heading_deviation = np.array(heading_deviation)
    time = np.array(time)
    
    # Calculate angular difference between consecutive heading deviations
    # Use arctan2 to handle circular wraparound properly
    d_heading_dev = np.diff(heading_deviation)
    d_heading_dev = np.arctan2(np.sin(d_heading_dev), np.cos(d_heading_dev))
    
    # Calculate time differences
    dt = np.diff(time)
    dt[dt == 0] = np.nan
    
    # Calculate rate of change
    heading_dev_change_vals = d_heading_dev / dt
    
    # Create output array with same length as input
    heading_dev_change = np.full(len(heading_deviation), np.nan)
    heading_dev_change[1:] = heading_dev_change_vals
    
    # Optional smoothing
    if smooth_window > 0:
        valid_mask = ~np.isnan(heading_dev_change)
        if np.sum(valid_mask) > 0:
            heading_dev_change[valid_mask] = scipy.ndimage.gaussian_filter1d(
                heading_dev_change[valid_mask], sigma=smooth_window)
    
    return heading_dev_change
```

## Process All Trials

```python
def process_all_trials_angular_velocity(df, conditions=[
        'all_light',
        'all_dark',
        'searchToLeverPath_light',
        'searchToLeverPath_dark',
        'homingFromLeavingLever_light',
        'homingFromLeavingLever_dark',
        'atLever_light',
        'atLever_dark'
    ],
    min_speed=2.0
):
    """Process all trials and calculate angular velocity and heading deviation change."""
    all_data = {}
    
    for condition in conditions:
        print(f"\nProcessing {condition}...")
        condition_df = df[df.condition == condition].copy()
        condition_df['session_trial'] = condition_df['session'] + '_T' + condition_df['trial'].astype(str)
        unique_trials = condition_df['session_trial'].unique()
        print(f"Found {len(unique_trials)} trials")
        
        processed_data = []
        
        for trial_id in tqdm(unique_trials):
            trial_data = condition_df[condition_df['session_trial'] == trial_id].copy()
            trial_data = trial_data.sort_values('recTime')
            
            if len(trial_data) < 10:
                continue
            
            # Calculate kinematic variables
            heading = calculate_heading_from_position(trial_data['x'].values, trial_data['y'].values)
            angular_velocity = calculate_angular_velocity(heading, trial_data['recTime'].values)
            speed = calculate_speed(trial_data['x'].values, trial_data['y'].values, trial_data['recTime'].values)
            
            # Calculate change in heading deviation
            heading_dev_change = calculate_heading_deviation_change(
                trial_data['mvtDirError'].values,
                trial_data['recTime'].values
            )
            
            trial_data['heading'] = heading
            trial_data['angular_velocity'] = angular_velocity
            trial_data['speed'] = speed
            trial_data['heading_dev_change'] = heading_dev_change
            
            # Filter for movement and valid data
            moving_mask = (speed >= min_speed) & ~np.isnan(angular_velocity) & ~np.isnan(heading_dev_change)
            trial_data_moving = trial_data[moving_mask]
            
            if len(trial_data_moving) > 0:
                processed_data.append(trial_data_moving)
        
        if len(processed_data) > 0:
            condition_combined = pd.concat(processed_data, ignore_index=True)
            all_data[condition] = condition_combined
            print(f"Total data points: {len(condition_combined)}")
        else:
            all_data[condition] = pd.DataFrame()
    
    return all_data
```

```python
# Process data
print("Processing all trials...")
all_data = process_all_trials_angular_velocity(dfAutoPI)

# Summary
for condition in all_data.keys():
    data = all_data[condition]
    if len(data) > 0:
        print(f"\n{condition}:")
        print(f"  Points: {len(data)}")
        print(f"  Ang vel range: [{data['angular_velocity'].min():.3f}, {data['angular_velocity'].max():.3f}] rad/s")
        print(f"  Heading dev change range: [{data['heading_dev_change'].min():.3f}, {data['heading_dev_change'].max():.3f}] rad/s")
        print(f"  Mean |ang vel|: {np.abs(data['angular_velocity']).mean():.3f} rad/s")
        print(f"  Mean |heading dev change|: {np.abs(data['heading_dev_change']).mean():.3f} rad/s")
```

## Visualization

```python
def plot_2d_histogram(data, condition):
    """Plot 2D histogram of angular velocity vs change in heading deviation."""
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    h, xedges, yedges = np.histogram2d(
        data['angular_velocity'], data['heading_dev_change'],
        bins=[50, 50], range=[(-3, 3), (-3, 3)])
    
    im = axes[0].imshow(h.T, origin='lower', aspect='auto', cmap='viridis',
                       extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
    axes[0].set_xlabel('Angular Velocity (rad/s)')
    axes[0].set_ylabel('Change in Heading Deviation (rad/s)')
    axes[0].set_title('2D Histogram')
    axes[0].axvline(x=0, color='white', linestyle='--', alpha=0.5)
    axes[0].axhline(y=0, color='white', linestyle='--', alpha=0.5)
    plt.colorbar(im, ax=axes[0], label='Count')
    
    h_log = np.log10(h + 1)
    im2 = axes[1].imshow(h_log.T, origin='lower', aspect='auto', cmap='viridis',
                        extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
    axes[1].set_xlabel('Angular Velocity (rad/s)')
    axes[1].set_ylabel('Change in Heading Deviation (rad/s)')
    axes[1].set_title('2D Histogram (Log Scale)')
    axes[1].axvline(x=0, color='white', linestyle='--', alpha=0.5)
    axes[1].axhline(y=0, color='white', linestyle='--', alpha=0.5)
    plt.colorbar(im2, ax=axes[1], label='Log10(Count + 1)')
    
    plt.suptitle(f'Angular Velocity vs Change in Heading Deviation - {condition}', fontsize=14, fontweight='bold')
    plt.tight_layout()
    return fig
```

```python
def plot_binned_analysis(data, condition, n_bins=10):
    """Plot binned analysis of angular velocity vs change in heading deviation."""
    from scipy.stats import mannwhitneyu
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Bin by angular velocity
    ang_vel_bins = np.linspace(-3, 3, n_bins + 1)
    bin_centers = (ang_vel_bins[:-1] + ang_vel_bins[1:]) / 2
    data['ang_vel_bin'] = pd.cut(data['angular_velocity'], bins=ang_vel_bins, labels=bin_centers)
    
    grouped = data.groupby('ang_vel_bin')
    mean_heading_dev_change = grouped['heading_dev_change'].mean()
    sem_heading_dev_change = grouped['heading_dev_change'].sem()
    count_per_bin = grouped.size()
    
    # Plot 1: Mean change in heading deviation vs angular velocity
    axes[0, 0].errorbar(mean_heading_dev_change.index.astype(float), mean_heading_dev_change.values, 
                       yerr=sem_heading_dev_change.values, fmt='o-', capsize=3, linewidth=2)
    axes[0, 0].axvline(x=0, color='black', linestyle='--', alpha=0.5)
    axes[0, 0].axhline(y=0, color='black', linestyle='--', alpha=0.5)
    axes[0, 0].set_xlabel('Angular Velocity (rad/s)')
    axes[0, 0].set_ylabel('Mean Change in Heading Deviation (rad/s)')
    axes[0, 0].set_title('Mean Change in Heading Deviation vs Angular Velocity')
    axes[0, 0].grid(True, alpha=0.3)
    
    # Plot 2: Data distribution
    axes[0, 1].bar(count_per_bin.index.astype(float), count_per_bin.values, 
                  width=(ang_vel_bins[1] - ang_vel_bins[0]) * 0.8, alpha=0.7)
    axes[0, 1].axvline(x=0, color='black', linestyle='--', alpha=0.5)
    axes[0, 1].set_xlabel('Angular Velocity (rad/s)')
    axes[0, 1].set_ylabel('Count')
    axes[0, 1].set_title('Data Distribution')
    axes[0, 1].grid(True, alpha=0.3)
    
    # Plot 3: Violin plot comparing left vs right turns
    left_data = data[data['angular_velocity'] > 0]['heading_dev_change'].dropna()
    right_data = data[data['angular_velocity'] < 0]['heading_dev_change'].dropna()
    
    violin_data = []
    labels = []
    if len(right_data) > 0:
        violin_data.append(right_data)
        labels.append(f'Right Turn (n={len(right_data)})')
        
    if len(left_data) > 0:
        violin_data.append(left_data)
        labels.append(f'Left Turn (n={len(left_data)})')
    
    if len(violin_data) > 0:
        parts = axes[1, 0].violinplot(violin_data, positions=range(len(violin_data)), 
                                       showmeans=True, showmedians=True)
        axes[1, 0].set_xticks(range(len(labels)))
        axes[1, 0].set_xticklabels(labels)
        axes[1, 0].axhline(y=0, color='black', linestyle='--', alpha=0.5)
        axes[1, 0].set_ylabel('Change in Heading Deviation (rad/s)')
        axes[1, 0].set_title('Distribution: Left vs Right Turns')
        axes[1, 0].grid(True, alpha=0.3, axis='y')
        
        # Perform Mann-Whitney U test and add significance marker
        if len(left_data) > 0 and len(right_data) > 0:
            statistic, p_value = mannwhitneyu(left_data, right_data, alternative='two-sided')
            
            # Determine significance marker
            if p_value < 0.001:
                sig_marker = '***'
            elif p_value < 0.01:
                sig_marker = '**'
            elif p_value < 0.05:
                sig_marker = '*'
            else:
                sig_marker = 'ns'
            
            # Add significance marker above the plot
            y_max = max(violin_data[0].max(), violin_data[1].max())
            y_min = min(violin_data[0].min(), violin_data[1].min())
            y_range = y_max - y_min
            y_pos = y_max + 0.1 * y_range
            
            axes[1, 0].plot([0, 1], [y_pos, y_pos], 'k-', linewidth=1.5)
            axes[1, 0].text(0.5, y_pos + 0.02 * y_range, sig_marker, 
                          ha='center', va='bottom', fontsize=12, fontweight='bold')
            axes[1, 0].text(0.5, y_pos + 0.08 * y_range, f'p = {p_value:.4f}', 
                          ha='center', va='bottom', fontsize=9)
    
    # Plot 4: Absolute values
    data['abs_ang_vel'] = np.abs(data['angular_velocity'])
    data['abs_heading_dev_change'] = np.abs(data['heading_dev_change'])
    abs_bins = np.linspace(0, 3, n_bins // 2 + 1)
    abs_bin_centers = (abs_bins[:-1] + abs_bins[1:]) / 2
    data['abs_ang_vel_bin'] = pd.cut(data['abs_ang_vel'], bins=abs_bins, labels=abs_bin_centers)
    abs_grouped = data.groupby('abs_ang_vel_bin')
    abs_mean = abs_grouped['abs_heading_dev_change'].mean()
    abs_sem = abs_grouped['abs_heading_dev_change'].sem()
    
    axes[1, 1].errorbar(abs_mean.index.astype(float), abs_mean.values,
                       yerr=abs_sem.values, fmt='o-', capsize=3, linewidth=2, color='purple')
    axes[1, 1].set_xlabel('|Angular Velocity| (rad/s)')
    axes[1, 1].set_ylabel('|Change in Heading Deviation| (rad/s)')
    axes[1, 1].set_title('Absolute Values')
    axes[1, 1].grid(True, alpha=0.3)
    
    plt.suptitle(f'Binned Analysis - {condition}', fontsize=14, fontweight='bold')
    plt.tight_layout()
    return fig
```

## Generate Plots

```python
def produce_two_figures(data, condition):
    """Generate and save both figure types for a condition."""
    fig1 = plot_2d_histogram(data, condition)
    plt.show()
    fig1.savefig(os.path.join(PROJECT_DATA_PATH, 'results', f'angular_velocity_heading_change_2d_{condition}.png'), 
                 dpi=300, bbox_inches='tight')

    fig2 = plot_binned_analysis(data, condition)
    plt.show()
    fig2.savefig(os.path.join(PROJECT_DATA_PATH, 'results', f'angular_velocity_heading_change_binned_{condition}.png'), 
                 dpi=300, bbox_inches='tight')
```

```python
conditions_to_process = [
    'all_light',
    'all_dark',
    'searchToLeverPath_light',
    'searchToLeverPath_dark',
    'atLever_light',
    'atLever_dark',
    'homingFromLeavingLever_light',
    'homingFromLeavingLever_dark'
]

for condition in conditions_to_process:
    if len(all_data[condition]) > 0:
        produce_two_figures(all_data[condition], condition)
```

## Statistical Analysis

```python
# Correlation analysis
print("\n" + "="*80)
print("STATISTICAL ANALYSIS: Angular Velocity vs Change in Heading Deviation")
print("="*80)

for condition in conditions_to_process:
    if len(all_data[condition]) > 0:
        data = all_data[condition]
        corr, pval = stats.pearsonr(data['angular_velocity'], data['heading_dev_change'])
        print(f"\n{condition}:")
        print(f"  Correlation (Pearson r): {corr:.4f}")
        print(f"  P-value: {pval:.6f}")
        
        # Compare left vs right turns
        left = data[data['angular_velocity'] > 0]
        right = data[data['angular_velocity'] < 0]
        if len(left) > 0 and len(right) > 0:
            t_stat, p_val = stats.ttest_ind(left['heading_dev_change'], right['heading_dev_change'])
            print(f"  Left vs Right t-test: t={t_stat:.3f}, p={p_val:.6f}")
            print(f"  Mean change (Left): {left['heading_dev_change'].mean():.4f} rad/s")
            print(f"  Mean change (Right): {right['heading_dev_change'].mean():.4f} rad/s")
```

## Summary

This analysis examines the relationship between turning rate (angular velocity) and the rate of error accumulation/correction (change in heading deviation).

### Key Interpretations:

**Positive correlation:**
- Turning in one direction → error changing in the same "direction"
- Could indicate turning causes error to accumulate

**Negative correlation:**
- Turning in one direction → error changing in opposite "direction" 
- Could indicate corrective turning behavior

**No correlation:**
- Turning rate is independent of error correction rate
- May suggest open-loop turning behavior

### Important Note:
Unlike the original analysis (angular velocity vs absolute heading deviation), this analysis examines **rates of change** for both variables. Both angular velocity and change in heading deviation are linear variables (rad/s), making Pearson correlation appropriate.

The biological meaning is different:
- **Original**: Does current error level relate to turning?
- **This analysis**: Does rate of error change relate to turning rate?
